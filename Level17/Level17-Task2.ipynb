```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv("D:/Mall_Customers.csv")
print(df.columns)
```

```python
features =  ['Age', 'Annual Income (k$)','Spending Score (1-100)']
X = df[features]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)
print(X_pca[:5])
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c='blue', marker='o')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('3D PCA of Customer Segmentation Data')
plt.show()
```

```python
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio for each component:", explained_variance_ratio)
cumulative_variance = explained_variance_ratio.cumsum()
print("Cumulative Explained Variance Ratio:", cumulative_variance)
```
